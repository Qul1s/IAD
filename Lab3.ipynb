Обрахувати – дисперсію, коваріацію та кореляцію по факторах. Зробити порівняльний аналіз між факторами.


```python
j=0
cov = np.zeros(18)
for i in np.arange(4, 22):
    factor = LED[i]
    cov[j] = LifeExData['Life expectancy'].cov(LifeExData[factor])
    print("Коваріація Life expectancy з ",LED[i],":",cov[j])
    j+=1
```

```python
j=0
corCoef = np.zeros(18)
for i in np.arange(4, 22):
    factor = LED[i]
    corCoef[j]= LifeExData['Life expectancy'].corr(LifeExData[factor])
    print("Кореляція",LED[i],corCoef[j])
    j+=1
```

```python
j=0
var = np.zeros(18)
for i in np.arange(4, 22):
    factor = LED[i]
    var[j] = LifeExData[factor].var()
    print("Дисперсія",LED[i],var[j])
    j+=1
```

```python
Fact = ['Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles','BMI','under-five deaths','Polio','Total expenditure','Diphtheria','HIV/AIDS','GDP','Population','thinness  1-19 years','thinness 5-9 years','Income composition of resources','Schooling']
allData = zip(Fact,cov,corCoef,var)
rezCol = ['Фактор','Коваріація','Кореляція','Дисперсія']
Rez = pd.DataFrame(data=allData, columns = rezCol)
Rez
```

```python
sb.catplot(Rez, y='Кореляція', hue='Фактор')
```

```python
sb.catplot(Rez, y='Коваріація', hue='Фактор')
```

```python
sb.catplot(Rez, y='Дисперсія', hue='Фактор')
```

Чи справді різні фактори прогнозування, які були обрані спочатку, впливають на очікувану тривалість життя? Які передбачувані змінні насправді впливають на тривалість життя?

```python
firstUsed = ['Adult Mortality','infant deaths','Alcohol','BMI','under-five deaths','Population']
print("Дані, що були обрані спочатку (без імунізації та індексів людського розвитку)")
for i in np.arange(0,6):
    result = LifeExData[firstUsed[i]].corr(LifeExData['Life expectancy'])
    if(-0.09<=result<=0 or 0<= result<=0.09): print(firstUsed[i], " Кореляція відсутня ", result)
    if(-0.3<result<=-0.1or 0.1<=result<0.3): print(firstUsed[i], " Низька ", result)
    if(-0.5<result<=-0.3or 0.3<=result<0.5): print(firstUsed[i], " Середня ", result)
    if(-1<=result<=-0.5 or 0.5<=result<=1): print(firstUsed[i], " Висока ", result)
allFactors = Fact
print(" ")
print("Дані, що насправді мають вплив на ОТЖ")
for i in np.arange(0, 18):
    result = LifeExData[allFactors[i]].corr(LifeExData['Life expectancy'])
    if(-0.5<result<=-0.3or 0.3<=result<0.5): print(allFactors[i], " Середня ", result)
    if(-1<=result<=-0.5 or 0.5<=result<=1): print(allFactors[i], " Висока ", result)
```

Чи повинна країна з нижчим значенням очікуваної тривалості життя (<65) збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя?


```python
under65 = LifeExData[LifeExData['Life expectancy']<65]
under65s = under65.sample(n = 20)
fig, ax = plt.subplots()
plt.title("Залежність СПЖ від витрат на охорону здоров'я")
ax.set_xticks(np.arange(30, 65))
ax = sb.scatterplot(under65s, x='Life expectancy', y ='Total expenditure', hue = 'Country', )
sb.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()

under65cor = under65['Total expenditure'].corr(under65['Life expectancy'])
print("Витрати на ОЗ мають залежність від СТЖ:")
if(-0.09<=under65cor<=0 or 0<= under65cor<=0.09): print(" Кореляція відсутня ", under65cor)
if(-0.3<under65cor<=-0.1or 0.1<=under65cor<0.3): print(" Низьку ", under65cor)
if(-0.5<under65cor<=-0.3or 0.3<=under65cor<0.5): print( " Середню ", under65cor)
if(-1<=under65cor<=-0.5 or 0.5<=under65cor<=1): print( " Високу ", under65cor)
```

Чи має очікувана тривалість життя позитивну чи негативну кореляцію зі звичками в їжі, способом життя, фізичними вправами, курінням, вживанням алкоголю тощо.


```python
habits = LifeExData
result = habits['Life expectancy'].corr(habits['Alcohol'])
if(result>=0): print("Alcohol має позитивну кореляцію")
else: print("Alcohol має негативну кореляцію")
result = habits['Life expectancy'].corr(habits['BMI'])
if(result>=0): print("BMI має позитивну кореляцію")
else: print("BMI має негативну кореляцію")
```
